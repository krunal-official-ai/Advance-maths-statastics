```python

student_data = pd.DataFrame({
    "Roll_No": [1, 2, 3, 4, 5],
    "Percentage": [78.89, 85.34, 92.80, 60.77, 74.23],
    "Grade": ["B", "A", "A", "C", "B"],
    "City": ["Surat", "Ahmedabad", "Surat", "Vadodara", "Rajkot"]
})

student_data
```


## 4️⃣ Data Representation – Multiple Graphs

Visualization helps identify:
- Trends
- Patterns
- Outliers


```python

import matplotlib.pyplot as plt

# Line chart
plt.figure()
plt.plot(sales_data['Month'], sales_data['Sales'])
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python

# Bar chart
plt.figure()
student_data['City'].value_counts().plot(kind='bar')
plt.title("Students by City")
plt.show()
```

```python

# Pie chart
plt.figure()
student_data['Grade'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Grade Distribution")
plt.ylabel("")
plt.show()
```


## 5️⃣ Population vs Sample

Population → All data  
Sample → Small subset


```python

population = np.arange(1, 10001)
sample1 = np.random.choice(population, size=50)
sample2 = np.random.choice(population, size=200)

print(np.mean(population))
print(np.mean(sample1))
print(np.mean(sample2))
```

```python

# Effect of sample size
print(np.std(sample1))
print(np.std(sample2))
```
