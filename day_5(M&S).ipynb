## 1️⃣ Random Variable

A random variable assigns **numerical values to random outcomes**.

### Types:
- Discrete (counts)
- Continuous (measurements)


```python

import pandas as pd
import numpy as np

customers_df = pd.DataFrame({
    "Day": range(1, 31),
    "Customers": np.random.randint(40, 120, 30)
})

customers_df.head()
```


## 2️⃣ Probability Distribution

Shows how probabilities are distributed over values of a random variable.


```python

pmf = customers_df["Customers"].value_counts().sort_index()
pmf
```

```python

import matplotlib.pyplot as plt

plt.figure()
pmf.plot(kind="bar")
plt.title("PMF of Daily Customers")
plt.show()
```


## 3️⃣ Mean (Expected Value) of Random Variable


```python

expected_value = sum(pmf.index * pmf.values)
expected_value
```

```python

customers_df["Customers"].mean()
```


## 4️⃣ Variance of Random Variable


```python

variance_value = sum(((pmf.index - expected_value) ** 2) * pmf.values)
variance_value
```

```python

customers_df["Customers"].var(), customers_df["Customers"].std()
```


## 5️⃣ Continuous Random Variable Example (Sales)


```python

sales = np.random.normal(loc=50000, scale=8000, size=1000)
np.mean(sales), np.var(sales)
```

```python

plt.figure()
plt.hist(sales, bins=30, density=True)
plt.title("Distribution of Daily Sales")
plt.show()
```
