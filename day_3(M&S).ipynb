## 1️⃣ Gaussian (Normal) Distribution

A **Normal Distribution** is:
- Bell-shaped
- Symmetric
- Mean = Median = Mode

### Real-world Examples:
- Heights
- Exam scores
- Measurement errors


```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
normal_data = np.random.normal(loc=70, scale=10, size=1000)
df = pd.Series(normal_data)

np.mean(normal_data), np.median(normal_data), df.mode()[0], np.std(normal_data)
```

```python

plt.figure()
plt.hist(normal_data, bins=30)
plt.title("Gaussian Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
sns.histplot(normal_data, kde=True)
```


## 2️⃣ Empirical Rule (68–95–99.7 Rule)

- 68% within ±1σ  
- 95% within ±2σ  
- 99.7% within ±3σ  

Screenshot 2026-01-28 at 4.19.07 PM.png


```python

mean = np.mean(normal_data)
std = np.std(normal_data)

within_1 = np.mean((normal_data >= mean-std) & (normal_data <= mean+std))*100
within_2 = np.mean((normal_data >= mean-2*std) & (normal_data <= mean+2*std))*100
within_3 = np.mean((normal_data >= mean-3*std) & (normal_data <= mean+3*std))*100

within_1, within_2, within_3
```


## 3️⃣ Skewness

Skewness measures **asymmetry**.

- Positive → Right tail longer  
- Negative → Left tail longer  


```python
np.random.seed(42)
normal_data = np.random.normal(loc=70, scale=10, size=1000)
df = pd.Series(normal_data)

mean = np.mean(normal_data)
median = np.median(normal_data)
mode = df.mode()[0]
std = np.std(normal_data)

# Pearson’s skewness coefficient formula to find skewness
skewness = (3 * (mean - median)) / std
skewness
```

```python

from scipy.stats import skew

np.random.seed(42)
positive_skew = np.random.exponential(scale=10, size=1000)
negative_skew = -positive_skew + 40

skew(positive_skew), skew(negative_skew)
```

```python

plt.figure()
plt.hist(positive_skew, bins=30)
plt.title("Positive Skewness")
plt.show()

plt.figure()
plt.hist(negative_skew, bins=30)
plt.title("Negative Skewness")
plt.show()
```


## 4️⃣ Kurtosis

Kurtosis measures **tailedness**.

- Mesokurtic → Normal  
- Leptokurtic → Heavy tails  
- Platykurtic → Flat  


Screenshot 2026-01-28 at 5.00.57 PM.png

```python
mean = np.mean(normal_data)
std = np.std(normal_data)

fourth_moment = np.mean((normal_data - mean)**4)
manual_kurtosis = fourth_moment / (std**4)

manual_kurtosis
```

```python

from scipy.stats import kurtosis

heavy_tail = np.random.standard_t(df=3, size=1000)

print(kurtosis(normal_data, fisher=False)) # Raw Kurtosis
print(kurtosis(normal_data)) # Excess Kurtosis = Raw Kurtosis - 3
print(kurtosis(heavy_tail))
```

```python

plt.figure()
plt.hist(normal_data, bins=30)
plt.title("Mesokurtic")
plt.show()

plt.figure()
plt.hist(heavy_tail, bins=30)
plt.title("Leptokurtic")
plt.show()
```
