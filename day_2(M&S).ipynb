## 1️⃣ Types of Statistics

### 🔹 Descriptive Statistics
- Summarizes raw data
- Easy to understand
- Used in reports & dashboards

### 🔹 Inferential Statistics
- Draw conclusions about population
- Uses probability
- Used in hypothesis testing & ML


```python

import numpy as np

data = [12, 15, 18, 20, 22, 25, 30]

print("Data:", data)
print("Minimum:", min(data))
print("Maximum:", max(data))
```


## 2️⃣ What is Descriptive Statistics?

Descriptive statistics helps to:
- Reduce large data
- Identify patterns
- Compare datasets

### Common Tools:
- Mean, Median, Mode
- Range, Variance, Standard Deviation
- Tables & Graphs


```python

import pandas as pd

df = pd.DataFrame({
                    "Salaries": [30000, 32000, 28000, 33000, 34000, 32000]
                  })

df
```


## 3️⃣ Measures of Central Tendency

### 🔸 Mean
Average value (affected by outliers)

### 🔸 Median
Middle value (robust to outliers)

### 🔸 Mode
Most frequent value


```python

salaries = df['Salaries']

print("Mean   :", np.mean(salaries))
print("Median :", np.median(salaries))
print("Mode   :", salaries.mode()[0])
```

```python

# Effect of outliers
salaries_with_outlier = salaries.tolist()
salaries_with_outlier.append(100000)

print("Mean with outlier:", np.mean(salaries_with_outlier))
print("Median with outlier:", np.median(salaries_with_outlier))
```

```python
s = pd.Series(salaries_with_outlier)
s.plot(kind="box")
```


## 4️⃣ Measures of Dispersion

Dispersion tells **how spread out data is**.

### 🔹 Range
Difference between max & min

### 🔹 Variance
Average squared deviation from mean

### 🔹 Standard Deviation
Square root of variance


```python
import pandas as pd

df = pd.DataFrame({
                    "Salaries": [30000, 32000, 28000, 33000, 34000, 32000]
                  })

df

print("Range:", np.max(salaries) - np.min(salaries))
print("Variance:", np.var(salaries),"Rupees")
print("Standard Deviation:", np.std(salaries),"Rupees")
```
